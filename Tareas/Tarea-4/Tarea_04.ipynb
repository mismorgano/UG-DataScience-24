```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X_train, y_target = iris.data[:, 2:], iris.target
```

```python
from numpy.linalg import norm
def Riesgo(X, y, C):
    return np.sum(norm(X - C[y])**2)
```

```python
riesgos = []

for k in range(2, 11):
    est = KMeans(k)
    est.fit(X_train)
    riesgos.append(Riesgo(X_train, est.labels_, est.cluster_centers_))
```

```python
plt.plot(range(2, 11), riesgos)
```

Usando el **elbow method** podemos notar que el numero de clustes es 3,
aunque 4 tampoco parece mal. Esta ambigüedad es una de las debilidades de
este método.
